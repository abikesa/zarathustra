# Moral Sentiments

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph (DAG)
G = nx.DiGraph()

# Add nodes and edges based on the neuron structure
G.add_edges_from([(1, 4), (2, 4), (3, 4), (4, 5), (5, 6)])

# Define positions for each node
pos = {1: (0, 2), 2: (1, 2), 3: (2, 2), 4: (1, 1), 5: (1, 0), 6: (1, -1)}

# Labels to reflect parts of a neuron
labels = {
    1: 'Duty', 
    2: 'Reward vs. Punishment', 
    3: 'Actions', 
    4: 'Custom',
    5: 'Utility', 
    6: 'Philosophy'
}

# Define node colors
node_colors = ['lemonchiffon', 'paleturquoise', 'mistyrose', 'thistle', 'lightgreen', 'lightsalmon']

# Create a figure and axis with a specified size
fig, ax = plt.subplots(figsize=(10, 8))  # Adjust width and height as needed

# Draw the nodes and edges first with turquoise edges
nx.draw(G, pos, with_labels=False, node_size=2000, node_color=node_colors, ax=ax, arrows=True, edge_color='turquoise')

# Draw the labels with the color of the previous node
for i, (node, (x, y)) in enumerate(pos.items()):
    previous_color = node_colors[i - 1] if i > 0 else node_colors[-1]  # Shift colors
    ax.text(x, y, labels[node], fontsize=12, ha='center', va='center', color=previous_color)

# Set the title
plt.title("Moral Sentiments")

# Display the plot
plt.show()
````

_Two Fractals Considered in [Wealth of Nations](https://www.gutenberg.org/cache/epub/3300/pg3300-images.html)_. In one fractal, the strategic resource inputs are the worker, capital, and land. In a sense, the owners `strategy` is to hire workers, the `payoff` is improvement in value of capital stock, and the `equilibrium` is land ownership. This equilibrium has been the basis of virtually all wars throughout history to today. Whether a future might emerge where the equilibrium at stake is platform ownership by Alphabet, Amazon, Apple, Meta, Microsoft, Elon, or Nvidia (AMEN!) is yet to be seen. For now land & energy are dearly needed to support the intelligence that will drive these platforms of the future. The other fractal begins with the `strategic` input, the `payoff` after taxes and input costs are deducted, and the `equilibrium` of the market place, which is dependent on household & business inputs as well as governments policies insofar as they affect taxes. [Andrei Shleifer's](https://scholar.google.com/citations?user=RB-RKykAAAAJ&hl=en) top-two most cited papers focuses on `inputs` (corporate governance) & `taxes/system` (law & finance). There's perhaps no room in the 21st century for the discussion of `frontier` territories like the wild west, where the equilibrium is adversarial and colonial in tone. But Andrei Shleifer's third most cited paper investigates cooperate ownership around the world, where countries are on a continuum from the wild frontiers to bustling capitals of the west. Taken together, these three themes are color-coded to ease mapping onto Dante's allegory: `paradiso`, `limbo`, and `inferno`.
